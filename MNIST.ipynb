```python
import sys
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
%matplotlib inline

from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.model_selection import train_test_split
```

```python
from sklearn.datasets import load_digits
mnist_ds = load_digits()
```

```python
images_and_labels = list(zip(mnist_ds.images, mnist_ds.target))
for index, (image, label) in enumerate(images_and_labels[:8]):
    plt.subplot(2,4, index + 1)
    plt.axis('off')
    plt.imshow(image, cmap=plt.cm.gray_r, interpolation='nearest')
    plt.title('Training: %i' % label)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(mnist_ds.data,mnist_ds.target,test_size=0.2)
print('Training data and target sizes: \n{}, {}'.format(X_train.shape,y_train.shape))
print('Test data and target sizes: \n{}, {}'.format(X_test.shape,y_test.shape))
```

```python
from sklearn.linear_model import LogisticRegression


class_logistic = LogisticRegression(max_iter=200)
class_logistic.fit(X_train, y_train)

y_pred = class_logistic.predict(X_test)
print("Accuracy of model = %2f%%" % (accuracy_score(y_test, y_pred )*100))
```
